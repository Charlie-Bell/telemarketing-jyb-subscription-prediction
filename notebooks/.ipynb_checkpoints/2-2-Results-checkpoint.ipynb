# Predicting Subscription Numbers of JYB Telemarketing Dataset: Results
<em>"The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed."</em> - <a href="https://www.kaggle.com/datasets/aguado/telemarketing-jyb-dataset">Telemarketing JYB Dataset on Kaggle</a><br>

The problem is a simple Binary Classifcation problem, to determine whether a customer will subscribe based on collected data.
The dataset is comprised of:
- Bank client data.
- Previous contact data.
- Social and economic attributes.
<br><br>
<b>This notebook presents and discusses the results for the trained models using F1-Score as a metric for model evaluation.</b>

## Import Libaries and Data
Data 'folds' from K-Fold are loaded using pickle.<br>
The trained models 'models' are imported using pickle.

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import recall_score, ConfusionMatrixDisplay, confusion_matrix

import sys
sys.path.append("../src/")

import training

with open('../data/interim/data_semi_prepped.csv') as f:
    df = pd.read_csv(f)
    subscribed = (df[['subscribed']]=='yes')
    data = df.drop(columns=['subscribed'])

with open("../data/processed/folds.p", "rb") as f:
    folds = pickle.load(file=f)
    
with open("../models/classifiers.p", "rb") as f:
    models = pickle.load(file=f)
```

## Model Selection
The dataset is skewed, so a high accuracy would not necessarily indicate good performance, in fact if we assume all outputs are false, i.e. nobody would subscribe, then the accuracy would be 89%. Instead we use F1-Score as a measure of model performance due to the skewedness of the dataset, this uses an even combination of the precision and recall. We want both a high precision and a high recall, a higher recall being more important than precision since a False Positive is better than a False Negative. The F1-Score can be between 0-1, 1 being the best and 0 being the worst.<br>
<b>In summary we only want to consider models with an accuracy greater than 89% and a high F1-Score.</b>

```python
model_names = ['lr','dt','rf','xgb','cbc']
best_model = training.metrics(folds, models, model_names)
```

### Model performance:
The Random Forest classifier (rf) trained on fold 1 gives the highest acuracy, 99.2%, and F1-Score, 0.96, outperforming the other models by far.<br>
The Decision Tree classifier (dt) followers second, which makes sense seeing as Random Forest is an aggregate of Decision Trees.<br>
Boosting methods CatBoost and XGBoost underperform terribly for ensemble methods, likely due to poor default hyperparameter choices. With hyperparameter tuning these could potentially outperform the Random Forest classifier, but for now the Random Forest classifier gives a sufficient accuracy and F1-score.<br>
Logistic regression seems to provide a consistently low accuracy on each fold.<br>
<b>Based on the F1-Score metric, the Random Forest classifier is the optimal model for this binary classification problem.</b> 
<br>
## Improvements:
- 5 folds may be too many for the size of the dataset, and a 2- or 3-fold cross-validation may be more appropriate.
- Hyperparameter tuning should net huge improvements on the boosting methods.
- Some other form of scalings should be tested to reduce the sensitivity to outliers.
- Fold 4 seems to result in poor performance on each model, indicating that the data may not be split appropriately, and it may be worth trying other methods or random seeds.

```python
y_pred = best_model.predict(folds[1]['X_test'])
y_true = folds[1]['y_test']

tn_gain, fp_gain, fn_gain, tp_gain = 0, -1000, -20000, 1000
(tn, fp), (fn, tp) = confusion_matrix(y_true, y_pred)
conf_mat = ConfusionMatrixDisplay.from_predictions(y_true, y_pred, display_labels=['Not subscribed','Subscribed'])


roi = pd.DataFrame({
    'Result': ['True Negative', 'False Positive', 'False Negative', 'True Positive'],
    'Gain per Customer': [tn_gain, fp_gain, fn_gain, tp_gain],
    'Count': [tn, fp, fn, tp],
    'Gain ($)':[tn*tn_gain, fp*fp_gain, fn*fn_gain, tp*tp_gain],
    },
).set_index('Result')
```

```python
def feature_importance(data, model):
    feature_importances = pd.DataFrame({
        'Feature': data.columns,
        'Importance': best_model.feature_importances_
    })
    feature_importances = feature_importances.sort_values(by='Importance', ascending=False)[:20]

    plt.bar(x=feature_importances['Feature'], height=feature_importances['Importance'], color='#08AE8B')
    plt.title('Feature importances obtained from impurity', size=20)
    plt.xticks(rotation='vertical')
    plt.show()
    
feature_importance(data, best_model)
```

The top 4 most important features, providing the most information gain, are plotted for all positive examples in the dataset. This way we can see the distribution

```python
top_4 = ['age']#, 'nr_employed', 'housing', 'day_of_week']
for feature in top_4:
    data[subscribed['subscribed']==True][feature].plot.kde()#.hist()
    data[subscribed['subscribed']==False][feature].plot.kde()#.hist()
```

Due to the nature of random forests, it's not easy to determine what groups should be targeted for future a future ad campaign. Some kind of clustering would be useful to determine these groups or it can also be done with a single decision tree classifier and looking at the split thresholds providing the highest information gain.

```python
def plot_kde(feature, data, subscribed):
    for i in [False, True]:
        if i:
            data[subscribed['subscribed']==i][feature].rename(str('Subscribed')).plot.kde(legend=True, title=feature+' probability density')
        else:
            data[subscribed['subscribed']==i][feature].rename(str('Not subscribed')).plot.kde(legend=True, title=feature+' probability density')
```

```python
plot_kde('age', data, subscribed)
```

```python
plot_kde('nr_employed', data, subscribed)
```

```python
plot_kde('housing', data, subscribed)
```

```python
plot_kde('day_of_week', data, subscribed)
```
